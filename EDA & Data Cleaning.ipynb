```python
def clean_address(address):
    if isinstance(address, str):
        # Replace newlines with comma and space, then remove multiple spaces
        cleaned_address = address.replace('\n', ', ').strip()
        # Replace multiple spaces with a single space
        cleaned_address = ' '.join(cleaned_address.split())
        return cleaned_address
    return address

df1['Address'] = df1['Address'].apply(clean_address)
```

```python
providers = pd.read_csv("clean_providers_data.csv")
receivers = pd.read_csv("clean_receivers_data.csv")
food = pd.read_csv("food_listings_data.csv")
claims = pd.read_csv("claims_data.csv")
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

## **Univariate**

```python
# 1. Provider Type Distribution

fig = px.histogram(
    providers,
    x="Type",
    color="Type",
    title="Provider Type Distribution"
)

fig.update_layout(
    xaxis_title="Provider Type",
    yaxis_title="Count"
)

fig.show()
```

```python
# 2. Receiver Type Distribution

fig = px.pie(
    receivers,
    names="Type",
    title="Receiver Type Distribution",
    hole=0.4
)

fig.show()
```

```python
# 3. Food Type Distribution

fig = px.bar(
    food["Food_Type"].value_counts().reset_index(),
    x="Food_Type",
    y="count",
    title="Food Type Distribution"
)

fig.show()
```

```python
# 4. Meal Type Distribution

fig = px.pie(
    food,
    names="Meal_Type",
    title="Meal Type Distribution"
)

fig.show()
```

## **Bivariate**

```python
# 5. City vs Food Listings
city_food = food.groupby("Location").size().reset_index(name="Listings")

fig = px.bar(
    city_food,
    x="Location",
    y="Listings",
    color="Listings",
    title="City vs Food Listings"
)

fig.show()
```

```python
# 6. Provider Type vs Quantity

provider_qty = food.groupby("Provider_Type")["Quantity"].sum().reset_index()

fig = px.bar(
    provider_qty,
    x="Provider_Type",
    y="Quantity",
    color="Quantity",
    title="Provider Type vs Quantity"
)

fig.show()
```

```python
# 8. Meal Type vs Quantity

meal_qty = food.groupby("Meal_Type")["Quantity"].sum().reset_index()

fig = px.bar(
    meal_qty,
    x="Meal_Type",
    y="Quantity",
    color="Quantity",
    title="Meal Type vs Quantity"
)

fig.show()
```

## **Multivariate**

```python
# 9. City + Provider Type + Quantity

city_provider = food.groupby(
    ["Location", "Provider_Type"]
)["Quantity"].sum().reset_index()

fig = px.sunburst(
    city_provider,
    path=["Location", "Provider_Type"],
    values="Quantity",
    title="City + Provider Type + Quantity"
)

fig.show()
```

```python
# 10. Food Type + Meal Type + Quantity

food_meal = food.groupby(
    ["Food_Type", "Meal_Type"]
)["Quantity"].sum().reset_index()

fig = px.treemap(
    food_meal,
    path=["Food_Type", "Meal_Type"],
    values="Quantity",
    title="Food Type + Meal Type + Quantity"
)

fig.show()
```

```python
# 11. Provider + Claims + Quantity

provider_claims = food.merge(
    claims,
    on="Food_ID"
)

provider_claims = provider_claims.groupby(
    "Provider_ID"
).agg(
    Claims=("Claim_ID", "count"),
    Quantity=("Quantity", "sum")
).reset_index()

fig = px.scatter(
    provider_claims,
    x="Claims",
    y="Quantity",
    size="Quantity",
    title="Provider Claims vs Quantity"
)

fig.show()
```

```python
# 12. Receiver + Claims + Quantity

receiver_claims = claims.merge(
    food,
    on="Food_ID"
)

receiver_claims = receiver_claims.groupby(
    "Receiver_ID"
).agg(
    Claims=("Claim_ID", "count"),
    Quantity=("Quantity", "sum")
).reset_index()

fig = px.scatter(
    receiver_claims,
    x="Claims",
    y="Quantity",
    size="Quantity",
    title="Receiver Claims vs Quantity"
)

fig.show()
```

```python
# 13. Claim Status Distribution

fig = px.pie(
    claims,
    names="Status",
    title="Claim Status Distribution",
    hole=0.5
)

fig.show()
```

```python
# 14. Top 10 Receivers

top_receivers = claims.groupby(
    "Receiver_ID"
).size().reset_index(name="Claims")

top_receivers = top_receivers.sort_values(
    "Claims",
    ascending=False
).head(10)

fig = px.bar(
    top_receivers,
    x="Receiver_ID",
    y="Claims",
    color="Claims",
    title="Top 10 Receivers"
)

fig.show()
```

```python
# 15. Top 10 Providers

top_providers = food.groupby(
    "Provider_ID"
)["Quantity"].sum().reset_index()

top_providers = top_providers.sort_values(
    "Quantity",
    ascending=False
).head(10)

fig = px.bar(
    top_providers,
    x="Provider_ID",
    y="Quantity",
    color="Quantity",
    title="Top 10 Providers"
)

fig.show()
```

```python
total_food = food["Quantity"].sum()
total_providers = providers["Provider_ID"].nunique()
total_receivers = receivers["Receiver_ID"].nunique()
total_claims = claims["Claim_ID"].nunique()

print("Total Food:", total_food)
print("Total Providers:", total_providers)
print("Total Receivers:", total_receivers)
print("Total Claims:", total_claims)
```
